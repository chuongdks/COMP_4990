```python
import pandas as pd

import pandas as pd
import openai
import time

# Set your OpenAI API key
openai.api_key = "YOUR_API_KEY"

# Load your dataset
df = pd.read_csv('/mnt/data/PostHistoryFirstVersion.csv')

# Create empty lists to store the new improved stuff
new_titles = []
new_bodies = []
new_tags = []

# Go through each row
for index, row in df.iterrows():
    # Construct the prompt
    prompt = f"""
You are an assistant that improves Stack Overflow questions.

Original Title:
{row['Title']}

Original Body:
{row['Body']}

Original Tags:
{row['Tags']}

Please:
- Rewrite the Title to be more clear and professional.
- Rewrite the Body to be clearer and more detailed if needed.
- Adjust Tags if necessary.

Format your response as:
Title: <your improved title>
Body: <your improved body>
Tags: <your improved tags separated by commas>
"""
    
    # Send the prompt to OpenAI
    response = openai.ChatCompletion.create(
        model="gpt-4-turbo",  # or "gpt-3.5-turbo" if you want cheaper
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0.3,
        max_tokens=1000
    )
    
    # Parse the response
    reply = response['choices'][0]['message']['content']
    
    # Very basic parsing
    title_line = reply.split('Title:')[1].split('Body:')[0].strip()
    body_line = reply.split('Body:')[1].split('Tags:')[0].strip()
    tags_line = reply.split('Tags:')[1].strip()
    
    new_titles.append(title_line)
    new_bodies.append(body_line)
    new_tags.append(tags_line)

# Add new columns to the DataFrame
df['Improved_Title'] = new_titles
df['Improved_Body'] = new_bodies
df['Improved_Tags'] = new_tags

# Save to new CSV
df.to_csv('ImprovedQuestions.csv', index=False)
print("Improvement complete and saved!")
```
