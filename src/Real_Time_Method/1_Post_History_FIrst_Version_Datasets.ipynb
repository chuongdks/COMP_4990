```python
import pandas as pd
```

# LOAD THE DATA AND FILTER INTIAL QUESTION INFO

```python
# Step 1: Load the data
df = pd.read_csv('./QueryResults10.csv')

# Step 2: Filter only the Initial Title, Body, Tags
filter = df['Name'].isin(['Initial Title', 'Initial Body', 'Initial Tags'])
df = df[filter]

# print(df)
```

# Grouped by PostID and Extract each row Initial version

```python
# Step 3: Group by PostId
grouped = df.groupby('PostId')

# # Debug: Display the dataframe after group by PostID object
# for post_id, group in grouped:
#     print(f"Post ID: {post_id}")
#     print(group)
#     print("-" * 40)
    
# Step 4: For each PostId, extract Title, Body, and Tags
records = []

for post_id, group in grouped:
    # Dictionary for each row. Has PostID, Initial Title, Initial BOdy,Initial Tag
    post = {'PostId': post_id}
    
    for row in group.itertuples():
        if row.Name == 'Initial Title':
            post['Title'] = row.Text
        elif row.Name == 'Initial Body':
            post['Body'] = row.Text
        elif row.Name == 'Initial Tags':
            post['Tags'] = row.Text
    
    records.append(post)
    
# print(records)

# Step 5: Turn records List into DataFrame
final_df = pd.DataFrame(records)

# Step 6: Display final dataframe
# print(final_df)
from IPython.display import display
display(final_df)

# Step 7: Export to csv file
final_df.to_csv(path_or_buf='./PostHistoryFirstVersion.csv',
          index = False) # False index param means no index for export data
```
