```python
import pandas as pd
import difflib
```

```python
# load CSV data
file_path = "./QueryResults.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
# Group the Name attribute
groups  = df.groupby(['Name'])

# Display the GroupBy object
for name, group in groups:
    print(f"Group: {name}")
    print(group)
    print("-" * 40)
```

```python
# Group by Name attribute
groups  = df.groupby(['Name'])

# Store the difference between edits here
diffs = {}

for name, group in groups:
    # Sort by Creation Date Attribute for the Dataframe of the Group
    group_sorted = group.sort_values('CreationDate')    
    previous_text = None
    changes = []
    
    # 
    for index, row in group_sorted.iterrows():
        current_text = str(row['Text'])
        if previous_text is not None:
            # Generate a diff
            diff = difflib.unified_diff(
                previous_text.splitlines(), 
                current_text.splitlines(),
                lineterm='',
                n=3  # context lines
            )
            changes.append('\n'.join(diff))
        previous_text = current_text
    
    diffs[name] = changes

diffs
```
